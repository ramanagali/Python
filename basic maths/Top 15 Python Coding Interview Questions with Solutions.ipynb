1. Write a Python Program to print Prime Numbers between 2 numbers

```python
start = 1
end = 20

print('Approach1')
for i in range(start, end+1):
  #if i>1:
  for j in range(2,i):
    if(i % j==0):
      break
  else:
    print(i)

print('Approach2')
for num in range(start,end):
  if all(num % i != 0 for i in range(2, num)):
    print(num)
```

3. Write a sorting function without using the list.sort function (descendring Order)


```python
data_list = [2,5,7,9,6,1,3,54,34,36, 76,21,99]
new_list = []

for a in data_list:
  min = a
  for x in data_list:
    if x > min:
      min = x
  new_list.append(min)

print(new_list)
```

4. Write a Python program to print Fibonacci Series

---




```python
def Fib(n):
  if n == 0: return 0
  elif n ==1: return 1
  else: return Fib(n-1) + Fib(n-2)

for i in range(0,12):
  print(Fib(i))
```

10. Write a Python program to implement a Binary Search


```python
def binary_search(arr, x):
  low = 0
  high = len(arr) - 1
  mid = 0

  while low <= high:

    mid = (high + low) // 2

    # If x is greater, ignore left half
    if arr[mid] < x:
      low = mid + 1

    # If x is smaller, ignore right half
    elif arr[mid] > x:
      high = mid - 1

    # means x is present at mid
    else:
      return mid

  # If we reach here, then the element was not present
  return -1
 
 
# Test array
arr = [ 2, 3, 4, 10, 40 ]
x = 10
 
# Function call
result = binary_search(arr, x)

if result != -1:
    print("Element is present at index", str(result))
else:
    print("Element is not present in array")
```

15. Given a string as your input, delete any reoccurring character, and return the new string.
  

```python
def delete_recurring_char(str):
  s = set()
  output = ''
  for c in str:
    if c not in s:
      s.add(c)
      output+= c
  return output

delete_recurring_char('venkataramana')
```

16. Split a string number into Number with comma sepertor

```python
def reverse(string):
  string = "".join(reversed(string))
  return string

def format_integer(number, seperator=','):
  s = reverse(str(number))
  count = 0
  result = ''
  for char in s:
    count = count+1
    if count % 3 == 0:
      if len(s) == count:
        result = char + result
      else:
        result = seperator + char + result
    else:
        result = char + result
  return result

input = '12345678'
format_integer(input)
  
```
